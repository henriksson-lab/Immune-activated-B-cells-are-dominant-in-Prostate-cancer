```r
exp_vol <- repExplore(immdata$data, .method = "volume")
by_vec <- c("N_G", "N_M", "N_G", "N_M", "N_G", "N_M", "P_G", "P_M", "P_G", "P_M", "P_G", "P_M", "S_G", "S_M", "S_G", "S_M", "S_G", "S_M")
b <- vis(exp_vol, .by = by_vec, .test = F)
ggsave(filename = "clono_GM.png", plot = b, width = 5, height = 6)
b
```

```r
# Compute statistics and visualise them
# Chao1 diversity measure
div_chao <- repDiversity(immdata$data, "chao1")

# Hill numbers
div_hill <- repDiversity(immdata$data, "hill")

# D50
div_d50 <- repDiversity(immdata$data, "d50")

# Ecological diversity measure
div_div <- repDiversity(immdata$data, "div")

p1 <- vis(div_chao)
p2 <- vis(div_hill)
p3 <- vis(div_d50)
p4 <- vis(div_div)

p1
p2
p3
p4
```

```r
imm_ov1 <- repOverlap(immdata$data, .method = "public", .verbose = F)
imm_ov2 <- repOverlap(immdata$data, .method = "morisita", .verbose = F)
imm_ov3 <- repOverlap(immdata$data, .method = "jaccard", .verbose = F)
imm_ov4 <- repOverlap(immdata$data, .method = "overlap", .verbose = F)
imm_ov5 <- repOverlap(immdata$data, .method = "tversky", .verbose = F)
imm_ov6 <- repOverlap(immdata$data, .method = "cosine", .verbose = F)
```

```r
library("RColorBrewer")
p1 <- vis(imm_ov1, "heatmap2", .title = "", .text.size = 10, .color = colorRampPalette(brewer.pal(n = 7, name = "Blues"))(100))
p1 
ggsave(filename = "FIGURE_overlap_heavy.png", plot = p1, width = 4.8, height = 4.5, dpi = 600)
```
